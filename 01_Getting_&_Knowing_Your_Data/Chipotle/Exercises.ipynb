### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
import requests

download_url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv"
target_tsv_path = "chipotle_data.tsv"

response = requests.get(download_url)
response.raise_for_status()
with open(target_tsv_path, "wb") as f:
    f.write(response.content)
print("Download ready.")
```
